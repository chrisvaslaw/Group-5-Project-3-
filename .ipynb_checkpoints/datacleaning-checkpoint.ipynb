```python
# Dependencies and Setup
import os
import pandas as pd
import datetime
import numpy as np
import requests
import json
from config import gkey
```

```python
#read the data sets from resources directory
csv_path = "Resources/zillowdata.csv"
house_df = pd.read_csv(csv_path)
```

```python
#read data
house_df
```

```python
# Target city
target_city = ["New York, NY","United States","Los Angeles-Long Beach-Anaheim, CA","Chicago, IL","Dallas-Fort Worth, TX"]

# Build the endpoint URL
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_city, gkey)
```

```python
lat_col = []
lng_col = []
for city in house_df["RegionName"]:
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(city, gkey) 
    #print(city)
    geo_data = requests.get(target_url).json()
    #print(geo_data)
    lat = geo_data["results"][0]["geometry"]["location"]["lat"]
    lng = geo_data["results"][0]["geometry"]["location"]["lng"]
 
```

```python
house_df['Lat'] = lat
house_df['Lng'] = lng
house_df
```

```python
lat_col = []
lng_col = []
for city in house_df["RegionName"]:
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(city, gkey) 
    geo_data = requests.get(target_url).json()
    lat = geo_data["results"][0]["geometry"]["location"]["lat"]
    lng = geo_data["results"][0]["geometry"]["location"]["lng"]
# Print the json
# print(geo_data)
    lat_col.append(lat)
    lng_col.append(lng)
```
